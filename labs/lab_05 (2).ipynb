1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# grafico simple

# datos
x = df['month_number']
y = df['total_profit']

# grafico

# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar
plt.plot(
    x, # eje x
    y, # eje y
    label='linea', # etiquetado
    color="blue", # color
    linewidth=2 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.title("Grafico Lineal") # agregar titulo
plt.xlabel('Número de Mes') # nombre eje x
plt.ylabel('Profit Total') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

```python
# datos
np.random.seed(42)
x = df['month_number']
y = df['total_profit']
#s = np.random.rand(*x.shape) * 800 + 500

# grafico
fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    x, # eje x
    y, # eje y
    #s, # tamano de los puntos
    c="b", # color
    alpha=0.7, # intensidad color
    #marker=r'$\clubsuit$', # forma de los puntos
    label="Par de datos (X,Y)" # etiquetdo fijando posicion
)
plt.xlabel("Número de Mes")
plt.ylabel("Profit Total")
plt.legend(loc='upper left')
plt.title("Grafico de Dispersión") # agregar titulo
plt.show()
```
